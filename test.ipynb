```python
import torch
import numpy as np
```

```python
layer = torch.nn.Linear(1028, 256)
```

```python
def block_diagonal(shape:tuple, K):
    out = torch.zeros(shape)

    val = 1/(2*K+1)

    for i in range(out.shape[1]):
        a = max(0, i-K)
        b = min(out.shape[1]-1, i+K)
        out[a:b,i] = val
    
    return out

def init_weights(module, fn):
    with torch.no_grad():
        module.weight = torch.nn.Parameter(fn(module.weight.shape, K=2))
```

```python
init_weights(layer, block_diagonal)
```
